```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://disfold.com/indonesia/companies/'

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('table', class_ = 'striped')
```

```python
soup.find_all('th')
```

```python
columns_title = soup.find_all('th')
```

```python
data_table_title = [title.text.strip() for title in columns_title]
```

```python
data_table_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = data_table_title)
```

```python
df
```

```python
soup.find_all('tr')
```

```python
data_columns = soup.find_all('tr')
```

```python
for row in data_columns[1:8]:
    data_row = row.find_all('td')
    vidual_data_row = [data.text.strip() for data in data_row]
    
    #print(vidual_data_row)
    leght = len(df)
    df.loc[leght] = vidual_data_row
```
